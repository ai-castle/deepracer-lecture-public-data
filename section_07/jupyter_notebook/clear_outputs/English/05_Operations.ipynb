# **Operations**

## **1. Arithmetic Operators**

- `+` (Addition), `-` (Subtraction)
- `*` (Multiplication), `/` (Division)
- `//` (Floor Division), `%` (Modulus)
- `**` (Exponentiation)


**1.1. Addition and Subtraction**
- `+` (Addition), `-` (Subtraction)

```python
x = 35
y = 10

print("x+y :", x+y)  # Addition
print("x-y :", x-y)  # Subtraction
```

**1.2. Multiplication and Division**
- `*` (Multiplication), `/` (Division)

```python
x = 35
y = 10

print("x*y :", x*y)  # Multiplication
print("x/y :", x/y)  # Division
```

**1.3. Division by zero results in an error**

```python
x = 10
y = 0

print(10 / 0)
```

**1.4. Floor Division and Modulus**
- `//` (Floor Division), `%` (Modulus)

```python
# Quotient and Remainder

x = 11
y = 3

print('Quotient: ', x // y)
print('Remainder: ', x % y)
```

**1.5. Exponentiation**

```python
# Exponentiation

x = 9

print(x**2)
```

**1.6. Square root**

```python
# Square Root

x = 9

print(x**(1/2))
```

## **2. Comparison Operators**

- `<` : greater than
- `<=` : greater than or equal to
- `>` : less than
- `>=` : less than or equal to
- `==` : equal to

**2.1. Greater than and greater than or equal to**

```python
print(1 < 2)
```

```python
print(1 <= 2)
```

**2.2. Less than and less than or equal to**

```python
print(1 > 2)
```

```python
print(1 >= 1)
```

**2.3. Equal to**
- The equal sign (`=`) is used for assigning variables, while the double equal sign (`==`) is used for determining whether two values are equal.


```python
print(3 == 4)
```

```python
print(3 == 3)
```

## 3. Boolean Operators

- and, or, not

**3.1. and operator**
- If both are True, the result is True.
- If either is False, the result is False.

```python
print(True and True)
```

```python
print(True and False)
```

```python
print(False and False)
```

**3.2. and operator in inequality**

```python
print((1 < 2) and (3 < 5))
```

```python
print((1 < 2) and (3 > 5))
```

```python
print((1 > 2) and (3 > 5))
```

**3.3. or operator**
- If both are False, the result is False.
- If either is True, the result is True.

```python
print(True or True)
```

```python
print(True or False)
```

```python
print(False or False)
```

**3.4. or operator in inequality**

```python
print((1 < 2) or (3 < 5))
```

```python
print((1 < 2) or (3 > 5))
```

```python
print((1 > 2) or (3 > 5))
```

**3.5. not operator**
- not True == False
- not False == True


```python
print(not True)
```

```python
print(not False)
```

**3.6. Arithmetic with Booleans**
- Booleans can be used in arithmetic operations like integers or floats.
- In this case, `True` is treated as 1 and `False` is treated as 0.

```python
print(True + True)
```

```python
print(True + False)
```

```python
print(False + False)
```

## 4. Compound Assignment Operators

- `x += 1` : adds 1 to the variable x
- `x -= 1` : subtracts 1 from the variable x
- `x *= 3` : multiplies the variable x by 3
- `x /= 3` : divides the variable x by 3

**4.1. Adding and Assigning**
- `x += 1` : adds 1 to the variable x

```python
x = 3
x += 1

print(x)
```

**4.2. Subtracting and Assigning**
- `x -= 1` : subtracts 1 from the variable x

```python
x = 3
x -= 1

print(x)
```

**4.3. Multiplying and Assigning**
- `x *= 3` : multiplies the variable x by 3

```python
x = 9
x *= 3

print(x)
```

**4.4. Dividing and Assigning**
- `x /= 3` : divides the variable x by 3

```python
x = 9
x /= 3

print(x)
```

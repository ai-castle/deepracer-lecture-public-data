# **if 조건문**

- if 문은 조건에 따라서 실행 여부를 결정하는 제어문이다.



## **예시 1**
- 구조
  ```python
  if 조건:
      실행 코드
  ```
- 주의 : 조건문 끝에 콜론( : ) 을 붙여줘야 하며 하위 코드는 반드시 들여쓰기 (Tab 또는 Space 4번)을 해야 한다.

````python
x = 30

if x > 20 :
    print("20보다 큽니다.")
````

## **예시 2**
- 구조
  ```python
  if 조건1:
      실행 코드 1
  else :
      실행 코드 2
  ```

- else 는 if 조건문을 만족하지 않을 시 실행되는 코드 블럭을 의미한다.

````python
x = 10

if x > 20 :
    print("20보다 큽니다.")
else :
    print("20보다 크지 않습니다.")
````

## **예시 3**
- 구조
  ```python
  if 조건1:
      실행 코드 1
  elif 조건2:
      실행 코드 2
  else :
      실행 코드 3
  ```

- elif 는 if 조건문을 만족하지 않을 시 실행되는 조건문이다.
- if 및 elif의 어느 조건도 만족하지 않으면 else에 있는 코드 블럭이 실행된다.

````python
x = 15

if x > 20 :
    print("20보다 큽니다.")
elif x > 10 :
    print("20보다 크지 않고 10보다 큽니다.")
else :
    print("10보다 크지 않습니다.")
````


## **예시 4**

- 구조
  ```python
  if 조건1:
      실행 코드 1
  elif 조건2:
      실행 코드 2
  elif 조건3:
      실행 코드 3
  else :
      실행 코드 4
  ```

- elif의 개수에는 제한이 없다.

````python
x = 5

if x > 20 :
    print("20보다 큽니다.")
elif x > 10 :
    print("20보다 크지 않고 10보다 큽니다.")
elif x > 0 :
    print("10보다 크지 않고 0 보다 큽니다.")
else :
    print("0보다 작거나 같습니다.")
````

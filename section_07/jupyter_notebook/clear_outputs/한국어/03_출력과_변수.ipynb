## **2. 주석 달기**

- `# 주석내용`
- 주석에 코드 설명을 적어 놓으면 보기 편하다
- 주석처리 된 구문은 코드 실행에 적용되지 않는다.

**2.1. 우측 끝에 주석**

```python
print("Hello, World") # 인사 문구
print(1+2)  # 1과 2의 합을 계산
```

**2.2. 줄 전체에 주석**
- 줄 전체 주석 단축키 : ctrl + /

```python
# 인사 문구
print("Hello, World")

# 1과 2의 합을 계산
print(1+2)
```

## **3. 변수에 값 대입**

- `변수명 = 값`

**3.1. 변수 하나 만들기**

```python
# 변수 만들기
x = 1

print(x+2)
```

**3.2. 변수를 이용해서 또 다른 변수 만들기**

```python
# 변수 만들기
x = 1

# 변수에 숫자 더하여 새 변수 생성
y = x + 2

print(y)
```

**3.3. 변수 값 업데이트**

```python
# 변수 만들기
x = 1

# 자기 자신에 수 더하기
x = x + 1

print(x)
```

**3.4. 문자 이후에 숫자가 나오는 변수명은 가능**

```python
hello1 = "안녕하세요"
print(hello1)
```

**3.5. 숫자 이후에 문자가 나오는 변수명은 불가**

# **자료형**

## **1. 자료형 확인**

- `type(자료)` : 자료형을 확인
- 자료형에는 다음과 같은 종류들이 존재합니다.
  - 정수형 (int)
  - 실수형 (float)
  - 문자형 (str)
  - 불린형 (bool)
  - 리스트형 (list)
  - 딕셔너리형 (dict)

## **2. 정수형(int)과 실수형(float)**

**2.1. 정수형 자료 확인**

````python
print(type(10))
````

**2.2. 실수형 자료 확인**

````python
print(type(10.5))
````

**2.3. 정수 뒤에 점(.)을 붙이면 실수형으로 변환됩니다.**

````python
print(type(10.))
````

**2.4. 지수 표기법(Exponential Notation)**
- `aen` : $a \times 10^n$
- `1e3` : $1 \times 10^3$
- `3e-2` : $3 \times 10^{-2}$

````python
x = 1e3
print('1e3 = ', x)

y = 3e-2
print('3e-2 = ', y)
````

**2.5. `abs()` : 절댓값 함수**

````python
print(abs(-10))
````

````python
print(abs(5.5))
````

## **3. 문자형 (str)**

- 문자형 (string)은 보통 `"문자내용"` 또는 `'문자내용'` 과 같이 작은 따옴표(') 또는 큰 따옴표(")를 사용하여 생성합니다.

**3.1. 큰 따옴표로 문자형 생성**

````python
x = "안녕하세요"

print(x)
print(type(x))
````

**3.2. 작은 따옴표로 문자형 생성**

````python
y = '반갑습니다'

print(y)
print(type(y))
````

## **4. 불린형 (bool)**

- 불린형 (bool)은 `True` 또는 `False`와 같이 생성합니다.

**4.1. 불린형 자료 True 를 생성합니다**

````python
ttt = True

print(ttt)
print(type(ttt))
````

**4.2. 불린형 자료 False를 생성합니다.**

````python
fff = False

print(fff)
print(type(fff))
````

## **5. 리스트형 (list)**
- 리스트형(list)은 인덱스 번호로 값을 호출하는 구조의 자료형입니다.
- 리스트를 생성할 때는 `[원소0, 원소1, 원소2, ...]` 와 같이 생성합니다.

**5.1. 리스트를 생성**

````python
x = [10,20,30,40]

print(x)
print(type(x))
````

**5.2. 리스트의 원소 호출**
- 리스트의 원소를 호출할 때는 `리스트명[인덱스번호]`와 같이 호출합니다.
- 리스트형의 인덱스는 정방향의 경우 0부터 1씩 증가하고 역방향의 경우 -1 부터 1씩 감소합니다.

````python
x = [10,20,30,40]

# 인덱스를 이용하여 리스트의 원소를 호출합니다.
print("0번 인덱스 원소 : ", x[0])
print("1번 인덱스 원소 : ", x[1])
print("2번 인덱스 원소 : ", x[2])
print("3번 인덱스 원소 : ", x[3])
````

**5.3. 범위를 벗어난 인덱스로 리스트의 원소를 호출하면 에러가 발생**

````python
x = [10,20,30,40]

print(x[4])
````

**5.4. 역방향 인덱스를 이용하여 리스트의 원소를 호출**

- 역방향의 경우 인덱스는 -1 부터 1씩 감소.

````python
x = [10,20,30,40]

# 역방향 인덱스를 이용하여 리스트의 원소를 호출합니다.
print("-1번 인덱스 원소 : ", x[-1])
print("-2번 인덱스 원소 : ", x[-2])
print("-3번 인덱스 원소 : ", x[-3])
print("-4번 인덱스 원소 : ", x[-4])
````

**5.5. 범위를 벗어난 인덱스로 리스트의 원소를 호출하면 에러가 발생**

````python
x = [10,20,30,40]

print(x[-5])
````

**5.6. len() 함수**
- `len(리스트 자료)` : 리스트의 원소 개수를 반환합니다.

````python
x = ['a','b','c']

print(len(x))
````

**5.7. 슬라이싱(Slicing)**
- 슬라이싱(slicing)은 리스트에서 특정 범위의 요소들을 추출하는 것을 말합니다.
- `리스트[m:n]` : m번 인덱스 이상, n번 인덱스 미만의 원소들을 가져오기

````python
x = ['a','b','c','d','e','f']

# 2번 인덱스 이상 4번 인덱스 미만 추출
x[2:4]
````

**5.8. 슬라이싱에서 끝 인덱스 생략**
- `리스트[m:]` : m번 인덱스 이상의 모든 원소들을 가져오기

````python
x = ['a','b','c','d','e','f']

# 2번 인덱스 이상 추출
print(x[2:])
````

**5.9. 슬라이싱에서 시작 인덱스 생략**
- `리스트[:n]` : n번 인덱스 미만의 모든 원소들을 가져오기

````python
x = ['a','b','c','d','e','f']

# 4번 인덱스 미만 추출
print(x[:4])
````

## **6. 딕셔너리형(dict)**

- 딕셔너리형(dict)은 key 값으로 value 를 호출하는 구조의 자료입니다.
- 딕셔너리를 생성할 때는 {key_A : value_A , key_B : value_B, ...} 와 같이 생성합니다.

**6.1. 딕셔너리를 생성**

````python
x = {'a':1, 'b':2, 'c':3}

print(x)
print(type(x))
````

**6.2 딕셔너리 자료에서 value 호출**
- 딕셔너리의 value를 호출할 때는 딕셔너리[key] 와 같이 호출합니다.

````python
x = {'a':1, 'b':2, 'c':3}

# 딕셔너리의 key 값으로 value를 호출합니다.
print("a :", x['a'])
print("b :", x['b'])
print("c :", x['c'])
````

**6.3 딕셔너리에 없는 key 호출하면 에러 발생**

````python
x = {'a':1, 'b':2, 'c':3}

# 딕셔너리에 정의되지 않은 key를 호출하면 에러가 발생합니다.
print(x['d'])
````

## **7. 자료형 변환**

- `float(자료)` : 실수형으로 변환
- `int(자료)` : 정수형으로 변환
- `str(자료)` : 문자형으로 변환

**7.1. float(정수형)**

````python
# 정수형을 실수형으로 변환
x = 10
y = float(x)

print(y)
print(type(y))
````

**7.2. float(문자형)**

````python
# 문자을 실수형으로 변환
x = '36.5'
y = float(x)

print(y)
print(type(y))
````

**7.3. 숫자가 아닌 문자를 실수형으로 변환하면 에러 발생**

````python
x = 'hello'
y = float(x)
````

**7.4. float(불린형)**
- `True`는 실수로 변환하면 0이 됩니다.
- `False`는 실수로 변환하면 0이 됩니다.

````python
# 불린형을 실수형으로 변환
ttt = float(True)
fff = float(False)

print("float(True) :", ttt)
print("float(False) :", fff)
````

**7.5. int(실수형)**
- 실수를 정수형으로 변환하면 소수점 자리를 모두 버립니다.

````python
print("int(1.7) :", int(1.7))
print("int(-3.6) :", int(-3.6))
````

**7.6. round() : 반올림 함수**
- `int()`함수는 단순히 소수점 자리를 모두 버립니다.
- 만약, 반올림을 하고자 한다면 `round()` 함수를 이용합니다.

````python
x = 1.6
y = round(x)

print(y)
````

**7.7. round() : 반올림 자리수 지정**
-  원하는 자릿수를 지정하여 반올림하고 싶다면 두 번째 인자로 자릿수를 지정할 수 있습니다. 

````python
# 반올림 자리수를 지정하여 반올림 가능합니다.
x = 1.6217

print("반올림 :", round(x))
print("소수점 첫째 자리에서 반올림 :", round(x,1))
print("소수점 둘째 자리에서 반올림 :", round(x,2))
print("소수점 셋째 자리에서 반올림 :", round(x,3))
````

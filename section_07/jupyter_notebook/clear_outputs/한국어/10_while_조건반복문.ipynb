# **while 조건반복문**

- "while"문은 조건이 참인 경우, 지정된 명령을 반복하는 반복문입니다.
- while 문의 구조는 다음과 같습니다.
  ```python
  while 조건:
      코드 블록
  ```
  - 위의 구조에서 `조건`은 반복 조건을 나타내는 불리언 값(True/False)입니다.
  - 위의 구조에서 `코드 블록`은 조건이 참일 때 실행될 코드입니다.

## **1. 기본 예시**

````python
count = 0

while count < 5:
    count += 1
    print(count)
    
````

## **2. 반복문 제어 : continue, break**

- `continue` : 다음 반복으로 넘어감
- `break` : 반복문을 종료

````python
count = 0

while count < 20:
    count += 1
    if count < 5 :
        continue
    elif count > 10 :
        break
    else :
        print(count)
````

## **3. 무한 반복문**

- `while True : ` 무한히 반복한다.
- 일반적으로 `break` 반복문 제어와 함께 사용한다.

````python
count = 0
while True :
    count +=1
    if count < 5 :
        print(count)
    else :
        break
````

# **연산**

## **1. 수 연산자**

- `+` (덧셈), `-` (뺄셈)
- `*` (곱하기), `/` (나누기)
- `//` (몫), `%` (나머지)
- `**` (거듭제곱)


**1.1. 덧셈과 뺄셈**
- `+` (덧셈), `-` (뺄셈)

```python
x = 35
y = 10

print("x+y :", x+y)  # 덧셈
print("x-y :", x-y)  # 뺄셈
```

**1.2. 곱하기와 나누기**
- `*` (곱하기), `/` (나누기)

```python
x = 35
y = 10

print("x*y :", x*y)  # 곱하기
print("x/y :", x/y)  # 나누기
```

**1.3. 0으로 나누면 에러 발생**

```python
x = 10
y = 0

print(10 / 0)
```

**1.4. 몫과 나머지**
- `//` (몫), `%` (나머지)

```python
# 몫과 나머지

x = 11
y = 3

print('몫: ', x // y)
print('나머지: ', x % y)
```

**1.5. 거듭제곱**

```python
# 제곱
x = 9

print(x**2)
```

**1.6. 거듭제곱근**

```python
# 제곱근
x = 9

print(x**(1/2))
```

## **2. 대소관계**

- `<` : 크다
- `<=` : 크거나 같다
- `>` : 작다
- `>=` : 작거나 같다
- `==` : 같다

**2.1. 크다, 크거나 같다**

```python
print(1 < 2)
```

```python
print(1 <= 2)
```

**2.2. 작다, 작거나 같다**

```python
print(1 > 2)
```

```python
print(1 >= 1)
```

**2.3. 같다**
- 등호(`=`)는 변수를 대입할 때 사용하고, 이중 등호 (`==`)는 두 값이 같은지 진위를 판정할 때 사용한다.

```python
print(3 == 4)
```

```python
print(3 == 3)
```

## **3. 불린 연산자**

- and (그리고), or (또는),  not (부정)

**3.1. and 연산자**
- 둘 다 True 이면 True
- 둘 중 어느 하나라도 False이면 False

```python
print(True and True)
```

```python
print(True and False)
```

```python
print(False and False)
```

**3.2. 부등식에서의 and 연산자**

```python
print((1 < 2) and (3 < 5))
```

```python
print((1 < 2) and (3 > 5))
```

```python
print((1 > 2) and (3 > 5))
```

**3.3. or 연산자**
- 둘 다 False 이면 False
- 둘 중 어느 하나라도 True이면 True

```python
print(True or True)
```

```python
print(True or False)
```

```python
print(False or False)
```

**3.4. 부등식에서의 and 연산자**

```python
print((1 < 2) or (3 < 5))
```

```python
print((1 < 2) or (3 > 5))
```

```python
print((1 > 2) or (3 > 5))
```

**3.5. 부정**
- not True == False
- not False == True

```python
print(not True)
```

```python
print(not False)
```

**3.6. 불린형의 수 연산**
- 불린형도 실수형 또는 정수형처럼 연산이 가능합니다.
- 이 때 `True`는 1로 `False`는 0으로 취급됩니다.

```python
print(True + True)
```

```python
print(True + False)
```

```python
print(False + False)
```

## **4. 연산과 대입 동시에**

- `x += 1` : x 변수에 1을 더함
- `x -= 1` : x 변수에 1을 뺌
- `x *= 3` : x 변수에 3을 곱함
- `x /= 3` : x 변수에 3을 나눔 

**4.1. 더하는 동시에 대입**
- `x += 1` : x 변수에 1을 더함

```python
x = 3
x += 1

print(x)
```

**4.2. 빼는 동시에 대입**
- `x -= 1` : x 변수에 1을 뺌

```python
x = 3
x -= 1

print(x)
```

**4.3. 곱하는 동시에 대입**
- `x *= 3` : x 변수에 3을 곱함

```python
x = 9
x *= 3

print(x)
```

**4.4. 나누는 곱하는 동시에 대입**
- `x /= 3` : x 변수에 3을 나눔 

```python
x = 9
x /= 3

print(x)
```

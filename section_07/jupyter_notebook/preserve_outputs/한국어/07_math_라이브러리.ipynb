# **math 라이브러리**

## **1. 라이브러리(Library) 란**

**1.1. 라이브러리란?**
  - "Library"는 프로그래밍 언어에서 이미 작성된 코드의 모음을 말합니다. 
  - 특정 기능을 수행하는 프로그램이나 기능을 미리 작성하여 저장하고, 다른 프로그램에서 이를 가져다 쓸 수 있는 것입니다.
  - 파이썬에서는 기본적으로 포함된 Library와 외부 Library 두 가지 종류의 Library가 있습니다. 
  - 기본적으로 포함된 Library는 파이썬을 설치할 때 자동으로 설치되는 Library이고, 외부 Library는 파이썬을 설치한 후에 추가적으로 설치하여 사용할 수 있는 Library입니다.
  - Library를 사용하면 프로그램을 작성하는 데 필요한 많은 기능을 간단하게 사용할 수 있어서, 프로그램을 개발하는데 많은 시간을 절약할 수 있습니다.

**1.2. math 라이브러리**
  - 파이썬 내장 라이브러리 중 하나로, 수학과 관련된 기능들을 제공한다.

**1.3. 라이브러리 사용방법**
  - 라이브러리를 불러오는 방법 :  `import 라이브러리명`  
    - (ex) `import math`
  - 라이브러리 내의 변수 등의 특정 기능을 호출할 때는, 라이브러리명 뒤에 점(.)을 붙여서 호출  
    - (ex) `math.log()` : math 라이브러리의 log함수

## **2. math 라이브러리**

**2.1. math 라이브러리 불러오기**

```python
import math
```

**2.2. 원주율(pi)과 자연상수(e)**
- `math.pi` : 원주율
- `math.e` : 오일러 자연상수

```python
print("원주율 :", math.pi) # 원주율
print("오일러 자연상수 :", math.e) # 오일러 자연상수
```

**2.3. math.degrees()**
- `math.degrees(라디안각)` : 라디안각을 60분법각으로 변환

```python
rad = math.pi  # 라디안각
deg = math.degrees(rad) # 라디안각을 60분법각으로 변환

print("radians :", rad)
print("degrees :", deg)
```

**2.4. math.radians()**
- `math.radians(60분법각)` : 60분법각을 라디안각으로 변환

```python
deg = 180  # 60분법각
rad = math.radians(deg) # 60분법각을 라디안각으로 변환

print("degrees :", deg)
print("radians :", rad)
```

**2.5. math.atan()**

- `math.atan(기울기)` : 기율기를 라디안각으로 변환


```python
rad = math.atan(1) # 기울기 1의 라디안각
deg = math.degrees(rad) # 라디안각을 60분법으로 변환

print("degree :", deg)
```

**2.6. math.atan2()**
- `math.atan2(y성분, x성분)` : 벡터를 라디안각으로 변환

```python
x = 1
y = 1

rad = math.atan2(y,x) # (1, 루트3) 벡터의 라디안각
deg = math.degrees(rad) # 라디안각을 60분법각으로 변환

print("degree :", deg)
```

**2.7. 삼각함수**

- `math.sin()` : 사인함수
- `math.cos()` : 코사인함수
- `math.tan()` : 탄젠트 함수

```python
# cos 함수
x = math.pi
y = math.cos(x)

print(y)
```

**2.8. 로그함수**

- `math.log()` : 자연로그
- `math.log2()` : 밑이 2인 로그
- `math.log10()` : 밑이 10인 로그

```python
# 자연로그
x = math.e
y = math.log(x)

print(y)
```

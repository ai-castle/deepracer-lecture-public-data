# **for 반복문**

- "for" 반복문은 파이썬에서 특정 구간 내에 있는 요소들을 순차적으로 순회하며 반복 처리할 때 사용하는 반복문입니다.
- 다음은 "for" 반복문의 기본적인 구조입니다:
  ```python
  for 변수 in 범위:
      수행할 문장
  ```
  - `변수`는 반복문을 순회할 때마다 값이 변경되는 변수입니다.
  - `범위`는 반복문을 순회할 요소들이 들어있는 구간입니다. 범위는 리스트, 튜플, 문자열, 딕셔너리 등의 시퀀스 자료형이나 range() 함수로 생성된 객체 등으로 지정할 수 있습니다.
  - `수행할 문장`은 반복문을 순회할 때마다 실행할 코드 블록입니다.

## **1. 리스트 원소 반복**

**1.1. 리스트 원소 반복**

````python
for i in ['a','b','c','d']:
    print(i)
````

## **2. range()를 활용한 반복**
- range() 함수는 for 반복문에서 사용되는 매우 유용한 함수 중 하나입니다. 
- range() 함수는 지정된 범위 내의 연속된 정수를 반환합니다.

**2.1. 0부터 5 미만까지 반복**

````python
for i in range(5):
    print(i)
````

**2.2. 2부터 5 미만까지 반복**

````python
for i in range(2,5):
    print(i)
    
````

**2.3. 2부터 10미만까지 3 간격으로 반복**

````python
for i in range(2,10,3):
    print(i)
````

## **3. 반복문 제어 : continue, break**

- `continue` : 다음 반복으로 넘어감
- `break` : 반복문을 종료

**3.1. continue**
- 다음 반복으로 넘어감

````python
for i in range(5) :
    if i < 2 :
        continue
    else :
        print(i)
        
````

**3.2. break**
- 반복문을 종료

````python
for i in range(5) :
    if i > 2 :
        break
    else :
        print(i)
        
````

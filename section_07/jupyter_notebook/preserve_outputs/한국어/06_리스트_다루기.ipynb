# **리스트 다루기**

## **1. 원소 추가하기**

- `리스트.append()`: 리스트의 맨 뒤에 새 요소를 추가할 때 사용합니다.
- `리스트.insert()`: 리스트의 특정 위치에 새 요소를 추가할 때 사용합니다.
- `리스트.extend()`: 리스트에 다른 리스트를 추가할 때 사용합니다.

**1.1. append() 매소드**
- `리스트.append()`: 리스트의 맨 뒤에 새 요소를 추가할 때 사용합니다.

```python
x = ['a', 'b', 'c']

x.append('d')
print(x)
```

**1.2. insert() 매소드**
- `리스트.insert()`: 리스트의 특정 위치에 새 요소를 추가할 때 사용합니다.

```python
x = ['a', 'b', 'c']

x.insert(1, 'd')
print(x)
```

**1.3. extend() 매소드**
- `리스트.extend()`: 리스트에 다른 리스트를 추가할 때 사용합니다.

```python
x = ['a','b','c']
y = ['d','e','f']

x.extend(y)
print(x)
```

## **2. 원소 제거**

- `리스트.remove(값)`: 리스트에서 특정 요소를 제거할 때 사용합니다.
- `리스트.pop(인덱스)`: 리스트에서 특정 요소를 제거하고, 그 값을 반환할 때 사용합니다.
- `del 리스트[인덱스]`: 리스트에서 특정 요소를 제거합니다.

**2.1. remove() 매소드**
- `리스트.remove(값)`: 리스트에서 특정 요소를 제거할 때 사용합니다.

```python
x = ['a', 'b', 'c', 'd']

x.remove('b')
print(x)
```

**2.2. pop() 매소드**
- `리스트.pop(인덱스)`: 리스트에서 특정 요소를 제거하고, 그 값을 반환할 때 사용합니다.

```python
x = ['a', 'b', 'c', 'd']

removed_value = x.pop(1)
print(removed_value)

print(x)
```

**2.3. del 명령어**
- `del 리스트[인덱스]`: 리스트에서 특정 요소를 제거합니다.

```python
x = ['a', 'b', 'c', 'd']

del x[-1]
print(x)
```

## **3. 리스트 연산**

- `리스트 + 리스트` : 두 리스트를 합침
- `리스트 * 정수` : 리스트를 반복

**3.1. 리스트1 + 리스트2**
- 두 리스트를 합칩니다.

```python
a = [1, 2, 3]
b = [4, 5, 6]

c = a + b
print(c)
```

**3.2. 리스트 * 정수**
- 리스트를 반복합니다.

```python
a = [1, 2, 3]

b = a * 3
print(b)
```

## **2. Adding comments**

- `# comment content`
- Adding comments to code explanations can make them easier to understand.
- Commented lines are not executed when the code is run.

**2.1. Comments at the end of a line**

```python
print("Hello, World") # greeting message
print(1+2)  # calculate the sum of 1 and 2
```

**2.2. Comments for an entire line**
- Shortcut key for commenting out an entire line: ctrl + /

```python
# greeting message
print("Hello, World")

# calculate the sum of 1 and 2
print(1+2)
```

## **3. Assigning values to variables**

- `variable_name = value`

**3.1. Creating a single variable**

```python
# creating a variable
x = 1

print(x+2)
```

**3.2. Creating another variable using variables**

```python
# creating a variable
x = 1

# add a number to the variable and create a new variable
y = x + 2

print(y)
```

**3.3. Updating variable values**

```python
# creating a variable
x = 1

# add a number to itself
x = x + 1

print(x)
```

**3.4. Variable names can have numbers after characters**

```python
hello1 = "Hello"
print(hello1)
```

**3.5. Variable names cannot start with numbers**

```python
1hello = "Hello"
```

**3.6. Variable names cannot have spaces**

```python
greet person = "Hello everyone"
```

**3.7. If you want to include spaces in variable names, use underscores (_) instead.**

```python
greet_person = "Hello everyone"
print(greet_person)
```

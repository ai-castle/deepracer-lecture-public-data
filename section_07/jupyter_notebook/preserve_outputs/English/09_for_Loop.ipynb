# **for Loop**

- The "for" loop is a loop used in Python to iterate through elements within a specific range and perform repetitive processing.
- Here is the basic structure of the "for" loop:
  ```python
  for variable in range:
      statement(s) to be executed
  ```
  - `variable` is a variable whose value changes each time the loop is iterated.
  - `range` is the range of elements to be iterated in the loop, which can be specified using a sequence type such as a list, tuple, string, dictionary, or an object created using the range() function.
  - `statement(s) to be executed` is the code block to be executed each time the loop is iterated.


## **1. Iterating through list elements**

**1.1. Iterating through list elements**

````python
for i in ['a','b','c','d']:
    print(i)
````

## **2. Iterating with range()**
- The range() function is one of the very useful functions used in for loops in Python.
- The range() function returns a sequence of continuous integers within the specified range.


**2.1. Iterating from 0 to less than 5**

````python
for i in range(5):
    print(i)
````

**2.2. Iterating from 2 to less than 5**

````python
for i in range(2,5):
    print(i)
    
````

**2.3. Iterating from 2 to less than 10 with a step of 3**

````python
for i in range(2,10,3):
    print(i)
````

## **3. Loop Control Statements: continue, break**

- `continue`: Skips to the next iteration
- `break`: Terminates the loop

**3.1. continue**
- Skips to the next iteration


````python
for i in range(5) :
    if i < 2 :
        continue
    else :
        print(i)
        
````

**3.2. break**
- Terminates the loop

````python
for i in range(5) :
    if i > 2 :
        break
    else :
        print(i)
        
````

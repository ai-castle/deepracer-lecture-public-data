# **while Loop**



- The "while" loop is a loop that repeats a specified command as long as the condition is True.
- The structure of the while loop is as follows:
  ```python
  while condition:
      code block
  ```
  - In the above structure, `condition` is a Boolean value (True/False) that represents the repetition condition.
  - In the above structure, `code block` is the code to be executed when the condition is True.


## **1. Basic Example**

````python
count = 0

while count < 5:
    count += 1
    print(count)
    
````

## **2. Loop Control Statements: continue, break**

- `continue`: Skips to the next iteration
- `break`: Terminates the loop

````python
count = 0

while count < 20:
    count += 1
    if count < 5 :
        continue
    elif count > 10 :
        break
    else :
        print(count)
````

## **3. Infinite Loop**

- `while True:` Repeats indefinitely.
- Generally used with `break` loop control.

````python
count = 0
while True :
    count +=1
    if count < 5 :
        print(count)
    else :
        break
````

# **Manipulating Lists**

## **1. Adding Elements**

- `list.append()`: Adds a new element to the end of the list.
- `list.insert()`: Adds a new element at a specific position in the list.
- `list.extend()`: Adds another list to the end of the list.

**1.1. append() method**
- `list.append()`: Adds a new element to the end of the list.


```python
x = ['a', 'b', 'c']

x.append('d')
print(x)
```

**1.2. insert() method**
- `list.insert()`: Adds a new element at a specific position in the list.

```python
x = ['a', 'b', 'c']

x.insert(1, 'd')
print(x)
```

**1.3. extend() method**
- `list.extend()`: Adds another list to the end of the list.

```python
x = ['a','b','c']
y = ['d','e','f']

x.extend(y)
print(x)
```

## **2. Removing Elements**

- `list.remove(value)`: Removes a specific element from the list.
- `list.pop(index)`: Removes a specific element from the list and returns its value.
- `del list[index]`: Removes a specific element from the list.

**2.1. remove() method**
- `list.remove(value)`: Removes a specific element from the list.

```python
x = ['a', 'b', 'c', 'd']

x.remove('b')
print(x)
```

**2.2. pop() method**
- `list.pop(index)`: Removes a specific element from the list and returns its value.

```python
x = ['a', 'b', 'c', 'd']

removed_value = x.pop(1)
print(removed_value)

print(x)
```

**2.3. del statement**
- `del list[index]`: Removes a specific element from the list.

```python
x = ['a', 'b', 'c', 'd']

del x[-1]
print(x)
```

## **3. List Operations**

- `list1 + list2`: Concatenates two lists.
- `list * integer`: Repeats the list.

**3.1. list1 + list2**
- Concatenates two lists.


```python
a = [1, 2, 3]
b = [4, 5, 6]

c = a + b
print(c)
```

**3.2. list * integer**
- Repeats the list a given number of times.


```python
a = [1, 2, 3]

b = a * 3
print(b)
```

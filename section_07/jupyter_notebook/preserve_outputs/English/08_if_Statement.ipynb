# **if Statement**

- An if statement is a control statement that determines whether or not to execute a code based on a condition.


## **Example 1**
- Syntax
  ```python
  if condition:
      code to execute
  ```

- Note: A colon(:) must be added at the end of the condition statement and the sub-code must be indented (Tab or 4 spaces).

````python
x = 10

if x > 20:
    print("It is greater than 20.")
````

## **Example 2**
- Syntax
  ```python
  if condition1:
      code to execute 1
  else:
      code to execute 2
  ```
- Note: If is the initial condition statement, and elif is the condition statement that starts when the above conditions are not satisfied.


````python
x = 100

if (0 < x) and (x < 20):
    print("It is between 0 and 20.")
else:
    print("It is less than or equal to 0, or greater than or equal to 20.")
````

## **Example 3**
- Syntax
  ```python
  if condition1:
      code to execute 1
  elif condition2:
      code to execute 2
  else:
      code to execute 3
  ```
  
- Note: else is the code block that is executed when no condition is satisfied. 
- If any of the conditions in the if or elif statement is satisfied, only the corresponding code block is executed, and the other code blocks are not executed.


````python
x = 3

if x > 20:
    print("It is greater than 20.")
elif x > 10:
    print("It is less than or equal to 20, but greater than 10.")
else:
    print("It is less than or equal to 10.")
````


## Example 4

- Structure
  ```python
  if condition1:
      code to be executed 1
  elif condition2:
      code to be executed 2
  elif condition3:
      code to be executed 3
  else:
      code to be executed 4
  ```
- There is no limit to the number of elif statements.


````python
x = -15

if x > 20:
    print("It is greater than 20.")
elif x > 10:
    print("It is less than or equal to 20, but greater than 10.")
elif x > 0:
    print("It is less than or equal to 10, but greater than 0.")
else:
    print("It is less than or equal to 0.")
````

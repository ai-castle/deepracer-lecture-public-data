# AWS DeepRacer Discount Factor Rate

- This code visualizes the extent to which the discount coefficient discounts future rewards in AWS DeepRacer.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

gamma_list = [0.999, 0.995, 0.99, 0.98, 0.97]
max_sec = 10
steps = np.arange(15*max_sec)
plt.figure(figsize=(10,5))

for gamma in gamma_list :
    plt.plot(gamma ** steps, label=gamma)

plt.xticks(np.arange(max_sec+1)*15, 
           pd.Series(np.arange(max_sec+1).astype('str')) + ' (' \
           + (pd.Series(15*np.arange(max_sec+1)).astype('str')) + ')'
           )
plt.yticks(np.arange(0,1.1,0.1))
plt.grid()
plt.legend()
plt.title("Discount Factor Rate")
plt.xlabel("Seconds (Steps)")
plt.ylabel("Discount")
plt.show()
```
